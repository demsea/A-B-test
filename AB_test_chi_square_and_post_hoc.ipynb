```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

## The Null Hypothesis and the Alternative Hypothesis.

* Null Theory -> all versions equal by CTR
* Altern theory -> all versions are not equal by CTR

```python
#confidence level is 95%
#alpha is significance level, treshhold to reject the null hypothesis
alpha = 0.05
```

## First we collect data and create contingency table


```python
# test a
url = "https://drive.google.com/file/d/1BygkqVOmx-UxkU9anhi-EYfBJ3hZerpT/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
test_a = pd.read_csv(path)
df_a = test_a.copy()

# test b
url = "https://drive.google.com/file/d/1QwmvSMXKrdCxOgUMCNEWINIIBXE5A68t/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
test_b = pd.read_csv(path)
df_b = test_b.copy()

# test c
url = "https://drive.google.com/file/d/1ycR3UwhU4hxCGfrRAj0z1a3CyC6_1OzR/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
test_c = pd.read_csv(path)
df_c = test_c.copy()

# test d
url = "https://drive.google.com/file/d/1-ov_nvf0kDXemtuezuqJcSIQbrMYaYHY/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
test_d = pd.read_csv(path)
df_d = test_d.copy()
```

```python
# How many clicks did each element get
eniac_a_clicks = df_a.loc[df_a["Name"]=="SHOP NOW", "No. clicks"].iloc[0]
eniac_b_clicks = df_b.loc[df_b["Name"]=="SHOP NOW", "No. clicks"].iloc[0]
eniac_c_clicks = df_c.loc[df_c["Name"]=="SEE DEALS", "No. clicks"].iloc[0]
eniac_d_clicks = df_d.loc[df_d["Name"]=="SEE DEALS", "No. clicks"].iloc[0]
```

```python
#How many visits did each page get
eniac_a_visits = 25326
eniac_b_visits = 24747
eniac_c_visits = 24876
eniac_d_visits = 25233
```

```python
#Number of visitors who didn't click on the button
eniac_a_no_click = eniac_a_visits - eniac_a_clicks
eniac_b_no_click = eniac_b_visits - eniac_b_clicks
eniac_c_no_click = eniac_c_visits - eniac_c_clicks
eniac_d_no_click = eniac_d_visits - eniac_d_clicks
```

```python
#Create a contingency table that shows the clicks and no clicks for each version of the website
clicks = [eniac_a_clicks, eniac_b_clicks, eniac_c_clicks, eniac_d_clicks]
noclicks = [eniac_a_no_click, eniac_b_no_click, eniac_c_no_click, eniac_d_no_click]

observed_results = pd.DataFrame(data = [clicks, noclicks],
                                columns = ["Version_A", "Version_B", "Version_C", "Version_D"],
                                index = ["Click", "No-click"])

observed_results
```

## Chi-square test

```python
chisq, pvalue, df, expected = chi2_contingency(observed_results)
```

```python
chisq
```

```python
pvalue
```

```python
#P is low, null must go
if pvalue > alpha:
  print("The p-value is larger than alpha, do not reject the null hypothesis")
else:
  print("The p-value is smaller than alpha, reject the null hypothesis.")
```

We reject the Null hypothesis and it means that at least one of the four different versions performed significantly better or worse than the others.

```python
# calculate the click through rate for each version
ctr_a = observed_results.loc["Click", "Version_A"] / sum(observed_results["Version_A"])
ctr_b = observed_results.loc["Click", "Version_B"] / sum(observed_results["Version_B"])
ctr_c = observed_results.loc["Click", "Version_C"] / sum(observed_results["Version_C"])
ctr_d = observed_results.loc["Click", "Version_D"] / sum(observed_results["Version_D"])

# create a table with all four versions and their click-through rates.
ctrs = pd.DataFrame({"version": ["A", "B", "C", "D"],
                     "CTR": [ctr_a, ctr_b, ctr_c, ctr_d]})

ctrs.sort_values("CTR", ascending=False)
```

## Results so far:
Though, our customers definetely prefer white color for the button, so we can either leave the existing one A 2,02% CTR , or swith it to C option 2,11%, we don't have a certain winner yet.
Pinpoint the best-performing version, post hoc tests were conducted to compare each version pairwise.

```python
#creating new tables for post hoc tests
click_ab = [eniac_a_clicks, eniac_b_clicks]
no_click_ab = [eniac_a_no_click, eniac_b_no_click]

a_b = pd.DataFrame([click_ab, no_click_ab],
                                  columns = ["A", "B"],
                                  index = ["click", "no-click"])
a_b
```

```python
click_ac = [eniac_a_clicks, eniac_c_clicks]
no_click_ac = [eniac_a_no_click, eniac_c_no_click]

a_c = pd.DataFrame([click_ac, no_click_ac],
                                  columns = ["A", "C"],
                                  index = ["click", "no-click"])
a_c
```

```python
click_ad = [eniac_a_clicks, eniac_d_clicks]
no_click_ad = [eniac_a_no_click, eniac_d_no_click]
a_d = pd.DataFrame([click_ad, no_click_ad],
                                  columns = ["A", "D"],
                                  index = ["click", "no-click"])
a_d
```

```python
click_bc = [eniac_b_clicks, eniac_c_clicks]
no_click_bc = [eniac_b_no_click, eniac_c_no_click]

b_c = pd.DataFrame([click_bc, no_click_bc],
                                  columns = ["B", "C"],
                                  index = ["click", "no-click"])
b_c
```

```python
click_bd = [eniac_b_clicks, eniac_d_clicks]
no_click_bd = [eniac_b_no_click, eniac_d_no_click]

b_d = pd.DataFrame([click_bd, no_click_bd],
                                  columns = ["B", "D"],
                                  index = ["click", "no-click"])
b_d
```

```python
click_cd = [eniac_c_clicks, eniac_d_clicks]
no_click_cd = [eniac_c_no_click, eniac_d_no_click]

c_d = pd.DataFrame([click_cd, no_click_cd],
                                  columns = ["C", "D"],
                                  index = ["click", "no-click"])
c_d
```

```python
# add correction=False, we need to add it if the table is 2*2 or the expected value bigger than 5
#apply Yates continuity correction when working with a 2x2 contingency table to adjust for continuity in discrete data.

#ab
chisq_ab, pvalue_ab, df_ab, expected_ab = chi2_contingency(a_b, correction=False)

#ac
chisq_ac, pvalue_ac, df_ac, expected_ac = chi2_contingency(a_c, correction=False)

#ad
chisq_ad, pvalue_ad, df_ad, expected_ad = chi2_contingency(a_d, correction=False)

#bc
chisq_bc, pvalue_bc, df_bc, expected_bc = chi2_contingency(b_c, correction=False)

#bd
chisq_bd, pvalue_bd, df_bd, expected_bd = chi2_contingency(b_d, correction=False)

#cd
chisq_cd, pvalue_cd, df_cd, expected_cd = chi2_contingency(c_d, correction=False)
```

When performing multiple pairwise comparisons in a post hoc test, the alpha level (significance level, typically set at 0.05) needs to be adjusted to control for the family-wise error rate (FWER)the probability of making at least one Type I error (false positive).

Each individual test must have a p-value < 0.0083 to be considered significant.

```python
bonf_alpha = alpha / 6
bonf_alpha
```

```python
pvalue_ab, pvalue_ac, pvalue_ad, pvalue_bc, pvalue_bd, pvalue_cd
```

```python
#pvalue is smaller than bonf_alpha for 5 out of 6, so we can reject Null theory.
#only for ac it is bigger, but because CTR rate is very similar and difference is not significant
```

Rule: if p-value is smaller than alpha we can reject Null theory (with 95% confidence)
